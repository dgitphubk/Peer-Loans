```python
zerocols = []
for col in numcols:
    if df[col].sum() == 0:
        zerocols.append(col)
```

Fix indicators

```python
pd.unique(df['policy_code'].values)
```

```python
df['policy_code'] = df['policy_code'].fillna(0)
```

```python
df['policy_code'] = df['policy_code'].astype(int)
```

```python
indicols = []
for col in df:
    if df[col].nunique() == 2:
        indicols.append(col)
```

```python
print(indicols)
```

```python
df[indicols].head()
```

```python
for col in indicols:
    print(pd.unique(df[col].values))
```

```python
for col in indicols:
    df[col] = df[col].fillna('NULL')
```

```python
df.head()
```

Null out remaining nans

```python
df.columns[df.isna().any()]
```

```python
nancols = df.columns[df.isna().any()]
for col in nancols:
    df[col] = df[col].fillna('NULL')
```

Save cleaned dataset

```python
df.to_csv('LoanStats_cleaned.csv', index=False)
```
